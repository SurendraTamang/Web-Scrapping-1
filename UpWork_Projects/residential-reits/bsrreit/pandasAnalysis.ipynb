```python
import pandas as pd
```

```python
today = pd.read_csv('./today.csv')
yesterday = pd.read_csv('./yesterday.csv')
```

```python
def get_both_rows(today, yesterday, which=None):
    """Find rows which are different between two DataFrames."""
    comparison_df = today.merge(yesterday, indicator=True, how='outer')
    if which is None:
        diff_df = comparison_df[comparison_df['_merge'] == 'both']
    else:
        diff_df = comparison_df[comparison_df['_merge'] == which]
    # diff_df.to_csv('data/diff.csv')
    return diff_df
```

```python
df = get_both_rows(today, yesterday)
```

```python
import pandas as pd
from datetime import date
td = pd.read_csv('./today.csv')
yd = pd.read_csv('./yesterday.csv')

def get_both_rows(today, yesterday, which=None):
    """Find rows which are different between two DataFrames."""
    comparison_df = today.merge(yesterday, indicator=True, how='outer')
    if which is None:
        diff_df = comparison_df[comparison_df['_merge'] == 'both']
    else:
        diff_df = comparison_df[comparison_df['_merge'] == which]
    # diff_df.to_csv('data/diff.csv')
    return diff_df

def get_left_rows(today, yesterday, which=None):
    """Find rows which are different between two DataFrames."""
    comparison_df = today.merge(yesterday, indicator=True, how='outer')
    if which is None:
        diff_df = comparison_df[comparison_df['_merge'] == 'left_only']
    else:
        diff_df = comparison_df[comparison_df['_merge'] == which]
    # diff_df.to_csv('data/diff.csv')
    return diff_df

df1 = get_both_rows(td, yd)
df2 = get_left_rows(td, yd)

df2.loc[df2['_merge'] == 'left_only', 'Timestamp'] = date.today().strftime("%d/%m/%Y")
df1.drop('_merge', inplace=True, axis=1)
df2.drop('_merge', inplace=True, axis=1)
final_df = df1.append(df2, ignore_index=True)
```
